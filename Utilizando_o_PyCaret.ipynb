```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP-on-Line-Graduation/main/datasets/arvore_titanic.csv')
df.head(15)
```

```python
# Target: nossa variável-alvo escolhida Survived.
# Target type: o tipo de variável-alvo, neste caso, uma categórica de 2 categorias, portanto binária.
# Original data shape: tamanho do dataset em linhas e colunas.
# Transformed data shape: alguma transformação gerou 2 colunas adicionais.
# Transformed train set shape: base treino, neste caso, com 623 linhas de um total de 891, 70%, portanto.
# Transformed test set shape: base treino, neste caso, com 268 linhas de um total de 891, 30%, portanto.
# Ordinal features: há 1 variável com dados ordinais, que é justamente a 1ª coluna PassengerId.
# Numeric features: quantidade de colunas numéricas[6].
# Categorical features: quantidade de varáveis categóricas.
# Rows with missing values: linhas com valores nulos.
# Preprocess: está ativada esta opção indicando que está acontecendo um pré-processamento dos dados.
# Imputation type: significa que há um tipo de preenchimento de valores simples, quando valores nulos são encontrados.
# Numeric imputation: preenchimento de valores em colunas numéricas[7].
# Categorical imputation: preenchimento de valores em colunas categóricas.
# Maximum on-hot encoding: diz respeito a um limitador na criação de dummies para variáveis numéricas[8].
# Fold Generator: diz respeito ao tipo de cross-validation utlizada.
# Log Experiment/Experiment Name: é possível registrar todas as tentativas realizadas (fazer um log do experimento).
from pycaret.classification import *
s = setup(df, target = 'Survived')
```

```python
best = compare_models()
```

```python
display(best)
```

```python
evaluate_model(best)
```

```python
dftest = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP-on-Line-Graduation/main/datasets/test_titanic.csv')
dftest.head(15)
```

```python
pr = predict_model(best, data=dftest)
pr
```
