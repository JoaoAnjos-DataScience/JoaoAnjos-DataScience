#Analisando os dados por Região

```python
import pandas as pd

# Converter os números de string para int
df_final['Ticket médio compra'] = pd.to_numeric(df_final['Ticket médio compra'], errors='coerce')
df_final['Numero de estrelas'] = pd.to_numeric(df_final['Numero de estrelas'], errors='coerce')
df_final['Probabilidade retorno compra'] = pd.to_numeric(df_final['Probabilidade retorno compra'], errors='coerce')
df_final['Idade'] = pd.to_numeric(df_final['Idade'], errors='coerce')

# Calculando a contagem de ocorrências de cada Região
sigla_counts = df_final['Regiao'].value_counts()

# Calculando o total de ocorrências
total_ocorrencias = sigla_counts.sum()

# Calculando as porcentagens
porcentagens = (sigla_counts / total_ocorrencias) * 100

# Calculando o ticket médio quebrado por Região
ticket_medio_por_sigla = df_final.groupby('Regiao')['Ticket médio compra'].mean()

# Calculando a média da quantidade de estrelas por Região
media_estrelas_por_sigla = df_final.groupby('Regiao')['Numero de estrelas'].mean()

# Calculando a atividades dos clientes  por Região ou seja o porcentagem de atividade daqueles clientes referentes a região do pais.
Atividades_Clientes = df_final.groupby('Regiao')['Numero de estrelas'].mean() /5 *50

# Calculando a média da quantidade de idade por Rgião
media_estrelas_por_idade = df_final.groupby('Regiao')['Idade'].mean()

# Calculando a média da probabilidade de retorno por Região
media_prob_retorno_por_sigla = df_final.groupby('Regiao')['Probabilidade retorno compra'].mean()

# Calculando a média da probabilidade de retorno por Região
probabilidadederetorno = df_final.groupby('Regiao')['Probabilidade retorno compra'].mean() /10 *100

# Calculando a forma de pagamento mais utilizada por Região
forma_pagamento_mais_utilizada = df_final.groupby('Regiao')['Forma de Pagamento'].apply(lambda x: x.mode().iloc[0])

# Criando um DataFrame para as porcentagens
porcentagens_df = pd.DataFrame({'Regiao': porcentagens.index, 'Porcentagem': porcentagens.values})

# Adicione uma coluna com o número de registros por região
porcentagens_df['Qtd.Registros'] = porcentagens_df['Regiao'].map(sigla_counts)
porcentagens_df['Porcentagem'] = porcentagens_df['Porcentagem'].map("{:.2f}%".format)

# Adicionando a coluna de Média de Idade por UF
porcentagens_df['Idade'] = porcentagens_df['Regiao'].map(media_estrelas_por_idade)

# Adicionando a coluna de ticket médio ao DataFrame de porcentagens
porcentagens_df['Ticket Médio'] = porcentagens_df['Regiao'].map(ticket_medio_por_sigla)
porcentagens_df['Ticket Médio'] = porcentagens_df['Ticket Médio'].map('R$ {:.2f}'.format)

# Adicionando a coluna de média de estrelas ao DataFrame de porcentagens
porcentagens_df['Média Estrelas'] = porcentagens_df['Regiao'].map(media_estrelas_por_sigla)

# Adicionando a coluna de média de estrelas ao DataFrame de porcentagens
porcentagens_df['Cliente é Regular'] = porcentagens_df['Regiao'].map(atividades_clientes)
porcentagens_df['Cliente é Regular'] = porcentagens_df['Cliente é Regular'].map('{:.2f}%'.format)

# Adicionando a coluna de média de probabilidade de retorno ao DataFrame de porcentagens
porcentagens_df['Média Probabilidade Retorno'] = porcentagens_df['Regiao'].map(media_prob_retorno_por_sigla)

#Coluna que descreve a % de retorno dos nossos clientes
porcentagens_df['% de Retorno do Cliente'] = porcentagens_df['Regiao'].map(probabilidadederetorno)
porcentagens_df['% de Retorno do Cliente'] = porcentagens_df['% de Retorno do Cliente'].map('{:.1f}%'.format)

# Adicionando a coluna de forma de pagamento mais utilizada ao DataFrame de porcentagens
porcentagens_df['Forma de Pagamento Mais Utilizada'] = porcentagens_df['Regiao'].map(forma_pagamento_mais_utilizada)

# Exibindo o DataFrame com as porcentagens, ticket médio, média de estrelas, média de probabilidade de retorno, forma de pagamento mais utilizada e número de registros
display(porcentagens_df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Faixas etárias (de 18 a 60 em intervalos de 4)
faixas_etarias = range(18, 60, 4)

# Crie uma nova coluna no DataFrame para a faixa etária
df_final['Faixa Etária'] = pd.cut(df_final['Idade'], bins=faixas_etarias)

# Crie o gráfico de barras por região
plt.figure(figsize=(16, 10))
plt.suptitle('Ticket Médio por Faixa Etária e Região', fontsize=16)

# Lista de regiões únicas
regioes_unicas = df_final['Regiao'].unique()

# Determine o número de subplots com base no número de regiões
num_subplots = len(regioes_unicas)

# Calcule o número de linhas e colunas para organizar os subplots
num_linhas = (num_subplots + 1) // 2
num_colunas = 2

# Itere sobre as regiões e crie subgráficos
for i, regiao in enumerate(regioes_unicas, start=1):
    plt.subplot(num_linhas, num_colunas, i)

    # Filtre os dados para a região atual
    dados_regiao = df_final[df_final['Regiao'] == regiao]

    # Agrupe os dados por faixa etária e calcule o ticket médio
    ticket_medio_por_faixa_etaria = dados_regiao.groupby('Faixa Etária')['Ticket médio compra'].mean().reset_index()

    # Crie um gráfico de barras para a região atual
    ax = sns.barplot(data=ticket_medio_por_faixa_etaria, x='Faixa Etária', y='Ticket médio compra', palette='viridis')
    ax.set_title(f'Região: {regiao}')
    ax.set_xlabel('Faixa Etária')
    ax.set_ylabel('Ticket Médio')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

    # Adicione os valores acima das barras
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')

# Ajuste o layout da figura
plt.tight_layout()

# Mostre a figura
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Lista de tipos únicos de forma de pagamento e siglas de estado
formas_de_pagamento_unicas = df_resultado_final['Forma de Pagamento'].unique()
siglas_estado_unicas = df_resultado_final['Sigla Estado'].unique()

# Função para atualizar as opções do segundo dropdown com base na Sigla Estado selecionada
def update_formas_de_pagamento(change):
    if change['type'] == 'change' and change['name'] == 'value':
        if sigla_estado_dropdown.value == 'Todas':
            forma_pagamento_dropdown.options = ['Todas'] + list(formas_de_pagamento_unicas)
        else:
            formas_para_sigla_estado = df_resultado_final[df_resultado_final['Sigla Estado'] == sigla_estado_dropdown.value]['Forma de Pagamento'].unique()
            forma_pagamento_dropdown.options = ['Todas'] + list(formas_para_sigla_estado)

# Função para atualizar o DataFrame com base nas seleções do usuário
def update_dataframe(change):
    if change['type'] == 'change' and change['name'] == 'value':
        if sigla_estado_dropdown.value == 'Todas':
            if forma_pagamento_dropdown.value == 'Todas':
                display(df_resultado_final)
            else:
                filtered_df = df_resultado_final[df_resultado_final['Forma de Pagamento'] == forma_pagamento_dropdown.value]
                display(filtered_df)
        else:
            if forma_pagamento_dropdown.value == 'Todas':
                filtered_df = df_resultado_final[df_resultado_final['Sigla Estado'] == sigla_estado_dropdown.value]
                display(filtered_df)
            else:
                filtered_df = df_resultado_final[(df_resultado_final['Sigla Estado'] == sigla_estado_dropdown.value) & (df_resultado_final['Forma de Pagamento'] == forma_pagamento_dropdown.value)]
                display(filtered_df)

# Criação do primeiro widget de filtro para Sigla Estado
sigla_estado_dropdown = widgets.Dropdown(
    options=['Todas'] + list(siglas_estado_unicas),
    description='Sigla Estado:',
    disabled=False
)

# Conectar a função de alteração ao evento de mudança no widget Sigla Estado
sigla_estado_dropdown.observe(update_formas_de_pagamento)

# Criação do segundo widget de filtro para Forma de Pagamento
forma_pagamento_dropdown = widgets.Dropdown(
    options=['Todas'],
    description='Forma de Pagamento:',
    disabled=False
)

# Conectar a função de alteração ao evento de mudança no widget Forma de Pagamento
forma_pagamento_dropdown.observe(update_dataframe)

# Exibir os widgets de filtro
display(sigla_estado_dropdown, forma_pagamento_dropdown)
```
