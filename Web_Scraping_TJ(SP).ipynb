```python
import requests
from bs4 import BeautifulSoup
import re
from tqdm import tqdm

# Lista para armazenar os números de processo e assuntos
numeros_processo = []
assuntos = []
foro = []
vara = []
juiz = []
reclamente = []
reclamado = []
distribuicao = []
area = []
valor = []
recla = []
reclaman = []

# Loop para percorrer as linhas do DataFrame com a barra de progresso
for index, row in tqdm(df.iterrows(), total=len(df), desc='Processando'):
    processo = row['Processo']  # Substitua 'Processo' pelo nome da coluna do DataFrame que contém os números de processo

    # URL da página atual
    tj_sp = f"https://esaj.tjsp.jus.br/cpopg/show.do?processo.codigo=2S001PXKE0000&processo.foro=100&processo.numero={processo}"

    # Definindo cabeçalhos personalizados
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        'Referer': 'https://www.google.com/'
    }

    # Fazendo a requisição HTTP com os cabeçalhos personalizados
    page = requests.get(tj_sp, headers=headers)

    # Verificando se a requisição foi bem-sucedida
    if page.status_code == 200:
        # Atribuindo características de texto aos dados
        pagina_do_processo = page.text

        # Analisando o HTML
        soup = BeautifulSoup(pagina_do_processo, 'html.parser')

        #Encontrar Reclamado
        tds = soup.find_all('td', {'class': 'nomeParteEAdvogado'})

        # Para cada elemento na lista tds
        for td1 in tds:
        # Realizar a limpeza dos caracteres de escape e unicode
        # Talvez no seu já esteja limpo, se for o caso, comente essa linha
          td_limpo1 = re.sub('[\t\n\xa0]', '', td1.text)
        #nomes1 = td_limpo1.split('Advogado:')
        reclamado.append(td_limpo1)

        #ENCONTRAR Advogado(a) e partes
        abutri = soup.find_all('td', {'class': 'nomeParteEAdvogado'})
        abutri = abutri[:1]
        # Para cada elemento na lista tds
        for td2 in abutri:
        # Realizar a limpeza dos caracteres de escape e unicode
        # Talvez no seu já esteja limpo, se for o caso, comente essa linha
          td_limpo2 = re.sub('[\t\n\xa0]', '', td2.text)
        #nomes2 = td_limpo2.split('Advogado:')
        reclamente.append(td_limpo2)

        # Encontrando os números de processo
        numeroprocesso = soup.find_all('span', {'id': 'numeroProcesso'})

        # Encontrando o assunto
        assunto = soup.find('span', {'id': 'assuntoProcesso'})

        # Encontrando foro
        fr = soup.find('span', {'id': 'foroProcesso'})

        # Encontrando vara
        vr = soup.find('span', {'id': 'varaProcesso'})

        # Encontrando Juiz
        jz = soup.find('span', {'id':'juizProcesso'})

        #Encontrando data de distribuição
        distribuicao_data = soup.find('div',{'id': 'dataHoraDistribuicaoProcesso'})


        # Encontrando a área do processo
        area_processo = soup.find('div', {'id': 'areaProcesso'})

        # Encontrando Valor da ação
        valor_acao = soup.find('div', {'id': 'valorAcaoProcesso'})




        # Limpando e armazenando os números de processo, assuntos, partes e advogados
        for processo in numeroprocesso:
            numeros_processo.append(processo.text)
            assuntos.append(assunto)
            foro.append(fr.text)
            vara.append(vr.text)
            juiz.append(jz)
            reclamado
            reclamente
            distribuicao.append(distribuicao_data.text)
            area.append(area_processo.text)
            valor.append(valor_acao)



# Exibindo os números de processo, assuntos, partes e advogados
for i in range(len(numeros_processo)):
    print("Número de processo:", numeros_processo[i])
    print("Assunto:", assuntos[i])
    print("Foro", foro[i])
    print("Vara", vara[i])
    print("Juiz", juiz[i])
    print("Reclamado", reclamado[i])
    print("Reclamante", reclamente[i])
    print("Data de Distribuição:", distribuicao[i])
    print("Área:", area[i])
    print("Valor da Ação:", valor[i])
    print()
```
